```python
df = data.drop(columns = ['created_at', 'venue_id', 'user_id', 'id'])
df.head()
```

```python
X = df.values
```

```python
clustering = MeanShift(bandwidth=0.1).fit(X)
```

```python
amount_near_centers = [0]*max(clustering.labels_+1)
for center in clustering.cluster_centers_:
    amount_near_centers[center] += 1
proper_centers_numbers = []
for i in range(len(amount_near_centers)):
    if amount_near_centers[i]>15:
        proper_centers_numbers.append(i)
```

```python
offices = [
    [33.751277, -118.188740],
    [25.867736, -80.324116],
    [51.503016, -0.075479],
    [52.378894, 4.885084],
    [39.366487, 117.036146],
    [-33.868457, 151.205134]
]
```

```python
def dist(A, B):
    x = (A[0] - B[0])**2
    y = abs(A[1] - B[1]) % 180
    y = y ** 2
    return x+y
```

```python
def min_dist(center, offices):
    out = np.inf
    for office in offices:
        d = dist(center, office)
        if out > d:
            out = d
    return out        
```

```python
num_dist_cord = []
for center in proper_centers:
    d = min_dist(clustering.cluster_centers_[center], offices)
    cord = clustering.cluster_centers_[center]
    num_dist_cord.append([center, d, cord])
```
