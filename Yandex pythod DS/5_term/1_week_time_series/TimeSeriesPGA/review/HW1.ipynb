```python
%pylab inline
import pandas as pd
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
import warnings
from itertools import product

def invboxcox(y,lmbda):
   if lmbda == 0:
      return(np.exp(y))
   else:
      return(np.exp(np.log(lmbda*y+1)/lmbda))
```

```python
df=pd.read_csv('WAG_C_M.csv', sep=';')
df.set_index('month', inplace=True)
```

Ряд не стационарен - явный тренд, нестабильная дисперсия, сезонность.
Сделаем преобразование Бокса-Кокса для стабилизации дисперсии:

```python
df['box'], lmbda = stats.boxcox(df['WAG_C_M'])
```

```python
#df['box']=np.log(df['WAG_C_M'])
df['box'].plot()
print("Оптимальный параметр преобразования Бокса-Кокса: %f" % lmbda)
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(df['box'])[1])
```

Подберем лаги для преобразования ряда к стационарному.

```python
df['d12']=df['box']-df['box'].shift(12)
```

```python
df['d12'].plot()
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(df['d12'][12:])[1])
```

```python
df['d12_1']=df['d12']-df['d12'].shift(1)
```

```python
df['d12_1'].plot()
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(df['d12_1'][13:])[1])
```

Теперь ряд стационарен

```python
sm.graphics.tsa.plot_acf(df['d12_1'][13:], lags=50)
```

q=1, Q=0

```python
sm.graphics.tsa.plot_pacf(df['d12_1'][13:], lags=50)
```

p=1, P=4

```python
ps = range(0, 2)
d=1
qs = range(0, 1)
Ps = range(0, 5)
D=1
Qs = range(0, 1)
```

```python
parameters = product(ps, qs, Ps, Qs)
parameters_list = list(parameters)
len(parameters_list)
```

```python
%%time
results = []
best_aic = float("inf")
warnings.filterwarnings('ignore')

for param in parameters_list:
    #try except нужен, потому что на некоторых наборах параметров модель не обучается
    try:
        model=sm.tsa.statespace.SARIMAX(df['box'], order=(param[0], d, param[1]), 
                                        seasonal_order=(param[2], D, param[3], 12)).fit(disp=-1)
    #выводим параметры, на которых модель не обучается и переходим к следующему набору
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    #сохраняем лучшую модель, aic, параметры
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
    
warnings.filterwarnings('default')
```

```python
result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending=True).head())
```

```python
print(best_model.summary())
```

Остатки:

```python
plt.figure(figsize(15,8))
plt.subplot(211)
best_model.resid[13:].plot()
plt.ylabel(u'Residuals')

ax = plt.subplot(212)
sm.graphics.tsa.plot_acf(best_model.resid[13:].values.squeeze(), lags=48, ax=ax)

print("Критерий Стьюдента: p=%f" % stats.ttest_1samp(best_model.resid[13:], 0)[1])
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(best_model.resid[13:])[1])
```

Остатки несмещены (подтверждается критерием Стьюдента) стационарны (подтверждается критерием Дики-Фуллера и визуально), неавтокоррелированы (подтверждается критерием Льюнга-Бокса и коррелограммой).
Посмотрим, насколько хорошо модель описывает данные:

```python
df['model'] = invboxcox(best_model.fittedvalues, lmbda)
plt.figure(figsize(15,7))
df['WAG_C_M'].plot()
df['model'].plot(color='r')
plt.ylabel('Wine sales')
pylab.show()
```
