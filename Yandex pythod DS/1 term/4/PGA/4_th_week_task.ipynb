```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import scipy
import statistics
import math
%matplotlib inline
```

```python
lam = 2
```

```python
exp_rv = sts.expon(lam) # выбрал распределение
sample = exp_rv.rvs(1000) # сделал выборку
x = np.linspace(0,6,1000)
pdf = exp_rv.pdf(x) # построил функцию распределения
```

```python
plt.hist(sample, normed=True)
plt.plot(x, pdf, label='dist dencity for exp')
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper right')
plt.show() # построил график функции распределения и гистограму
```

For this distribution Expectation and Variance are defined by the only feature $\lambda$ as lam

$\mathbb{E} = \lambda ^-1$, $\mathbb{D} = \lambda ^-2$

```python
lam = 1
exp_rv = sts.expon(scale = lam)
```

```python
ns = [5, 10, 50]
N = 1000 
avg_n = [] # список для средних

for n in ns:
    avg_n = []
    for i in range(N):
        sample = exp_rv.rvs(size=n) #получили выборку
        avg_n.append(statistics.mean(sample)) # взяли среднее
    plt.hist(avg_n, density=True) # добавили среднее
    print("E = 1, D = 1, n = %d" %n) # указал мат ожидание и дисперсию
    
    ##
    ## Выкладки с вычислением параметров распределения - строчка ниже
    ##
    norm_rv = sts.norm(lam, scipy.sqrt(1.0/(n * lam**2))) 
    x = np.linspace(0,6,1000)
    norm_pdf = norm_rv.pdf(x) 
    plt.plot(x, norm_pdf, label=f'{lam} * exp(- {lam}*x)')# построил функцию распределения
    legend = plt.legend(shadow=True, fontsize='medium', loc='upper left')
    plt.ylabel('$average$')
    plt.xlabel('$x$')
    plt.show()
```
